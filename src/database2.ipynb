```python
#https://qiita.com/saira/items/e08c8849cea6c3b5eb0c
import sqlite3
import pandas as pd
import os

dest = 'databases'
if not os.path.exists(dest):
    os.makedirs(dest)

# pandascsv
# csv1label, 2wind_speed, 3temperature, 4precipitation_amount, 
#5wave_height
df = pd.read_csv("csv-files/kuroshima.csv", header=None)

# csv filecolumn
# pandas1

df.columns = [
    "1.Date",
    "2.Average-temperature()",
    "3.Maximum-temperature()",
    "4.Minimum-temperature()",
    "5.Total-precipitation(mm)",
    "6.Daylight-hours(hours)",
    "7.Average-wind-speed(m/s)",
    "8.Maximum-wind-speed(m/s)",
    "9.Maximum-wind-speed(direction)",
    "10.Maximum-instantaneous-wind-speed(m/s)",
    "11.Maximum-instantaneous-wind-speed(wind-direction)",
    "12.Most-wind-direction(16-directions)",
    "13.Average-vapor-pressure(hPa)",
    "14.Average-local-pressure(hPa)",
    "15.Average-sea-level-pressure(hPa)",
    "16.Minimum-sea-level-pressure(hPa)",
    "17.Average-humidity(%)",
    "18.Minimum-relative-humidity(%)",
    "19.Average-cloud-cover(10-minutes-ratio)",
    "20.Weather-overview(noon:06:00-18:00)",
    "21.Weather-overview(Night:18:00-to-06:00-the-next-day)",
    "22.Wave-height(m)",
    "23.Cancellation-label"]

dbname = os.path.join(dest,'kuroshima.db')

conn = sqlite3.connect(dbname)
cur = conn.cursor()

# dbnamesamplecsvsql
# if_existsexpense
df.to_sql('data', conn, if_exists='replace')

# 1
select_sql = 'SELECT * FROM data'
for row in cur.execute(select_sql):
    print(row)

cur.close()
conn.close()
```
